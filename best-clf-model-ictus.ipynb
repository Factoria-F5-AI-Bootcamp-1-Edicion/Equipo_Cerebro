```python
df_ros['balance_method'] = 'ros'
df_rus['balance_method'] = 'rus'
df_smote['balance_method'] = 'smote'
df_smote_tomek['balance_method'] = 'smote_tomek'
df_cnn['balance_method'] = 'condensed_nearest'
df_near_miss['balance_method'] = 'near_miss'
#df_sample_weight['balance_method'] = 'sample_weight'
#df_balance_bagging['balance_method'] = 'balance_bagging'
```

```python
metricas_balanceo = pd.concat([df_ros, df_rus,df_smote , df_smote_tomek , df_cnn , df_near_miss ])
metricas_balanceo
```

```python
metricas_balanceo.reset_index(inplace=True)
metricas_balanceo
```

```python
metricas_balanceo.groupby(['balance_method', 'gridSearch'])['recall'].aggregate('mean').unstack()
```

```python
metricas_balanceo.pivot_table(index='balance_method', columns='gridSearch',
                    aggfunc={'acc_test':'mean', 'recall':'mean'})
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

metricas_balanceo.pivot_table('recall', index='balance_method',
                    columns='gridSearch', aggfunc='mean').plot(figsize=(10,6))

plt.tick_params(axis="x", length=6, width=2, colors='r', labelrotation=45.0)
plt.tick_params(axis="y", length=3, width=4, colors='b')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tight_layout()
plt.ylabel('recall')

plt.show()
#plt.gca().set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
#plt.ylabel('mean births by day');
```

```python
metricas_balanceo.pivot_table('acc_test', index='balance_method',
                    columns='gridSearch', aggfunc='mean').plot(figsize=(10,6))
                                                               
plt.tick_params(axis="x", length=6, width=2, colors='r', labelrotation=45.0)
plt.tick_params(axis="y", length=3, width=4, colors='b')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tight_layout()
plt.ylabel('accuracy (test)')
                                                               
plt.show()
```

```python
metricas_balanceo.pivot_table('overfitting %', index='balance_method',
                    columns='gridSearch', aggfunc='mean').plot(figsize=(10,6))

plt.tick_params(axis="x", length=6, width=2, colors='r', labelrotation=45.0)
plt.tick_params(axis="y", length=3, width=4, colors='b')
plt.ylabel('overfitting (%)')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tight_layout()

plt.show()
```
